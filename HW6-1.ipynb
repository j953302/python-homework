# 找三維圖中的最低點

輸入我們需要的資源

```python
%matplotlib inline

from ipywidgets import interact, fixed
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
import math 
```

先嘗試畫畫看
x :　-1 ~ 1
y :　-2~2
z = 2x^2+y 
的圖形

```python
x = [np.linspace(-1,1,50)]
y = [np.linspace(-2,2,50)]

X, Y =np.meshgrid(x,y)

def Znumber(X, Y):
    
    return 2*X**2+Y
for i in range(0,len(x)):
    Z = Znumber(X, Y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot a basic wireframe.
ax.plot_wireframe(X, Y, Z)
 
plt.show()
```

定義尋找z最小值的方程式

```python
def findMinimum(x,y):
    fig = plt.figure()

    ax = fig.add_subplot(111, projection='3d')
    xa = []
    ya = []
    za = []
    η = 0.01
    n = 0
    z = 2*x**2+y
    print("Iteration "+str(n)+": x = "+str(x)+", y= "+str(y)+", z ="+str(2*x**2+y))        
    while (2*x**2+y)-(2*(x-η * 4*x)**2+y)> 0 and -1<=x:
        n = n + 1
        x = x - η * 4*x
        xa.append(x)
        ya.append(y)
        za.append(2*x**2+y)
        print("Iteration "+str(n)+": x = "+str(x)+", y= "+str(y)+", z ="+str(2*x**2+y))
    while (2*x**2+y)-(2*(x)**2+y-η)> 0and -2<=y:
        n = n + 1
        y = y - η
        xa.append(x)
        ya.append(y)
        za.append(2*x**2+y)
        print("Iteration "+str(n)+": x = "+str(x)+", y= "+str(y)+", z ="+str(2*x**2+y))
    ax.plot_wireframe(X, Y, Z,color = "y")
    ax.scatter3D(xa,ya,za,color = "r")
    ax.set_zlabel("z")
    ax.set_ylabel("y")
    ax.set_xlabel("x")
    plt.show()
```

執行從點(-1,-1) 開始

```python
findMinimum(-1,-1)
```

加入互動的元素，馬上跑出我們要的圖

```python
interact(findMinimum,x=(-1.0,1.0),y =(-2.0,2.0))
```
